```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('default')
%matplotlib inline
```

```python
def f(x):
    
    return np.cosh(x)*np.cos(x)
```

```python
X = np.linspace(-2,2,100)
```

```python
output = [f(x) for x in X]
```

```python
ones = [1 for x in X]
minus_ones = [-1 for x in X]
```

```python
#plot
fig, axes = plt.subplots(1,1)

axes.plot(X,output)
axes.plot(X,ones)
axes.plot(X,minus_ones)


axes.set_xlabel(f'x', fontsize=20)
axes.set_ylabel(f'f(x)', fontsize=20);
plt.title(f'cosh(x)*cos(x)')
```
